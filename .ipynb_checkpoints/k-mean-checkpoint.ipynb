```python
import cx_Oracle
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
con = cx_Oracle.connect('i73190', 'abc*2020', '172.31.0.11/STYLEDB')
records = []
cur = con.cursor()
rows = cur.execute("Select  a.freq_a,a.freq_t, a.freq_gc From beegenes a")
for r in rows:
    records.append(r)
cur.close()
con.close()
X = np.array(records)
```

```python
kmeans = KMeans(n_clusters=7, init="random",n_init=10, max_iter=500, tol=1e-4, random_state=0).fit(X)
labels = kmeans.labels_
plt.figure(figsize=(14,14))
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=20, c='red', label ='Cluster 1',marker="o")
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=20, c='blue', label ='Cluster 2',marker="o")
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=20, c='aqua', label ='Cluster 3',marker="o")
plt.scatter(X[labels==3, 0], X[labels==3, 1], s=20, c='black', label ='Cluster 4',marker="o")
plt.scatter(X[labels==4, 0], X[labels==4, 1], s=20, c='purple', label ='Cluster 5',marker="o")
plt.scatter(X[labels==5, 0], X[labels==5, 1], s=20, c='magenta', label ='Cluster 6',marker="o")
plt.scatter(X[labels==6, 0], X[labels==6, 1], s=20, c='green', label ='Cluster 7',marker="o")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', label = 'Centroids',marker="*")
plt.title('K-Means')
plt.legend(loc="upper right")
plt.show()
```
